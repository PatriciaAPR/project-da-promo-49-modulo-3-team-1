*** 1.Factores de Retención y Rotación de Empleados: ***
     Para conocer quien se va y quien se queda.

 - attrition ->  ¿El empleado ha dejado la empresa? (Yes/No)
 - yearsatcompany -> ¿Cuántos años lleva en la empresa?
 - yearsincurrentrole -> ¿Cuánto tiempo lleva en su puesto actual?
 - yearssincelastpromotion -> ¿Hace cuánto fue su última promoción?
 - yearswithcurrmanager -> ¿Cuántos años lleva con el mismo jefe?

    ¿Por qué son importantes?
        -> Si hay empleados que se van tras pocos años, podemos analizar qué lo causa.
        -> Si la falta de promociones o cambio de jefe influye en la rotación.

*** 2.Factores de Satisfacción en el Trabajo: ***
    Estas columnas nos dicen si los empleados están felices o nos ayuda a hacernos una idea de su satisfacción.

 - jobsatisfaction -> Nivel de satisfacción con el trabajo (1-4).
 - environmentsatisfaction -> Nivel de satisfacción con el ambiente laboral.
 - relationshipsatisfaction -> ¿Se llevan bien con compañeros y jefes?
 - worklifebalance -> ¿Tienen equilibrio entre trabajo y vida personal?
 - jobinvolvement -> Nivel de compromiso del empleado en el trabajo
    
    ¿Por qué son importantes?
    -> Si los empleados con baja satisfacción son los que más se van.
    -> Identificar qué empresa tiene mejor ambiente y bienestar.

*** 3.Factores Económicos y Beneficios: ***
    Tenemos que ver cómo influye el salario en la satisfacción y retención.

-  salary -> Salario total anual.
-  monthlyincome -> Ingreso mensual.
-  hourlyrate -> Tarifa por hora.
-  percentsalaryhike -> Incremento porcentual de salario (%) en los últimos años.
-  stockoptionlevel -> Nivel de opciones sobre acciones asignadas.

    ¿Por qué son importantes?
    -> ¿Ganan más los empleados que se quedan?
    -> ¿Los aumentos de sueldo afectan la rotación?

*** 4.Factores de Carga Laboral y Crecimiento ***

- overtime -> ¿Trabajan horas extras? (Yes/No)
- trainingtimeslastyear -> ¿Cuántos entrenamientos/formacion han recibido en el último año?
- joblevel -> Nivel jerárquico del empleado en la empresa.
- numcompaniesworked -> ¿En cuántas empresas han trabajado antes?
- totalworkingyears -> Años totales de experiencia laboral.

     ¿Por qué son importantes?
    -> ¿Las personas con mucha carga laboral se van antes?
    -> ¿Los empleados con más formación tienen mejor satisfacción?

*** 5.Factores Personales y Demográficos ***

- age -> Edad del empleado.
- gender -> Género.
- maritalstatus -> Estado civil (Married, Single, Divorced).
- numberchildren -> Número de hijos.
- remoteWork -> ¿Trabaja en remoto? (Yes/No)
- distancefromhome -> ¿Vive muy lejos del trabajo?

     ¿Por qué son clave?
        -> ¿Los empleados más jóvenes cambian de trabajo más seguido?
        -> ¿El trabajo remoto influye en la satisfacción?
        -> ¿El tiempo de traslado al trabajo afecta la rotación?
